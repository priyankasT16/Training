```python
import folium
import os
import pandas as pd
```

```python
map = folium.Map(location=[38.58,-99.09],zoom_start=6,title="Stamen Terrain")
fg=folium.FeatureGroup(name='My map')


os.getcwd()
```

```python
html="""Volcano name:<br>
<a href="https://www.google.com/search?q=%%22%s%%22" target="_blank">%s</a><br>
Height : %s m
"""
```

```python
def color_producer(elevation):
    if elevation<1000:
        return 'green'
    elif elevation <= 3000 and elevation >= 1000:
        return 'orange'
    else:
        return 'red'
```

```python
data=pd.read_csv('C:/Users/priyanka.thakare/Desktop/Jupyter/code/app2/data/Volcanoes.txt')
lat=list(data['LAT'])
lon=list(data['LON'])
elv=list(data['ELEV'])
name=list(data['NAME'])
for i,j,k,nm in zip(lat,lon,elv,name):
    iframe=folium.IFrame(html=html % (nm,nm,k),width=200,height=100)
    fg.add_child(folium.CircleMarker(location=[i,j],popup=folium.Popup(iframe),radius=6,fill_color=color_producer(k),
                                     color='grey',fill_capacity=0.7))
map.add_child(fg)
map.save("Map1.html")
```
